# Red neuronal con compuerta AND

```python
import matplotlib.pyplot as plt
import numpy as np

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

Como vemos en la siguiente imagen, esta es una compuerta AND
<img src="and.png">

Entonces, para los datos que le pasaremos a la red neuronal utilizaremos estos datos de la siguiente forma:
* Los datos de entrada que se ven en la imagen, serán pasados como pares de datos (p.e. [0,0]) para el X de entrenamiento.
* Los datos de salida, serán pasados para el Y de entrenamiento.

```python
x_train = np.array([[0,0],[0,1],[1,0],[1,1]])
Y_train = np.array([[0],[0],[0],[1]])
```

El modelo de perceptrón multicapa MLP, se definiría de la siguiente manera, utilizando la función de activación ReLU y sigmoid

```python
kerass = keras.Sequential()
kerass.add(layers.Dense(2, input_dim = 2, activation = 'relu'))
kerass.add(layers.Dense(1, activation = 'sigmoid'))
kerass.compile(optimizer = keras.optimizers.Adam(0.01), loss='mean_squared_error', metrics = ['accuracy'])
```

```python
fit_history = kerass.fit(x_train, Y_train, epochs = 400, batch_size = 1)
loss_curve = fit_history.history['loss']
accuracy_curve = fit_history.history['accuracy']
plt.plot(accuracy_curve, label = 'Precisión')
plt.plot(loss_curve, label = 'Perdida')
plt.legend(loc = 'lower left')
plt.title('Resultado del Entrenamiento')
plt.show
```

Recuperamos bias and weights de la capa oculta:

```python
pesos_HL, sesgos_HL = kerass.layers[0].get_weights()
pesos_HL, sesgos_HL
```

Recuperamos bias and weights de la capa de salida:

```python
pesos_OL, sesgos_OL = kerass.layers[1].get_weights()
pesos_OL, sesgos_OL
```
