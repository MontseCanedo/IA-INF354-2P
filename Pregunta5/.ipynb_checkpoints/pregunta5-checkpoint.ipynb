```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
import seaborn as sns
%matplotlib inline
label_encoder = preprocessing.LabelEncoder()
```

```python
data = pd.read_csv('breast-cancer.csv')
data.head()
```

```python
data_cols = pd.read_csv('Cnames.csv')
data_col_list = data_cols['Column_values'].tolist()
na_values = ["?"]
data2 = pd.read_csv('breast-cancer.csv', header=None, names = data_col_list, na_values = na_values)
data2.head()
```

```python
data2.columns.values
data2.tail(60)
```

```python
data2["Node_caps"] = data2["Node_caps"].fillna(method="ffill")
```

```python
data2.tail(60)
```

```python
def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix = var_name)
    df = df.drop(var_name, axis = 1)
    df = pd.concat([df, dummy], axis = 1)
    return df
```

```python
encoder_class = label_encoder.fit_transform(data2['Class'])
data2['Class'] = encoder_class
```

```python
encoder_nodecaps = label_encoder.fit_transform(data2['Node_caps'])
data2['Node_caps'] = encoder_nodecaps
```

```python
encoder_irradiat = label_encoder.fit_transform(data2['Irradiat'])
data2['Irradiat'] = encoder_irradiat
```

```python
encoder_age = label_encoder.fit_transform(data2['Age'])
data2['Age'] = encoder_age
```

```python
encoder_tumor_size = label_encoder.fit_transform(data2['Tumor_size'])
data2['Tumor_size'] = encoder_tumor_size
```

```python
encoder_menopause = label_encoder.fit_transform(data2['Menopause'])
data2['Menopause'] = encoder_menopause
```

```python
encoder_invnodes = label_encoder.fit_transform(data2['Inv_nodes'])
data2['Inv_nodes'] = encoder_invnodes
```

```python
encoder_breast = label_encoder.fit_transform(data2['Breast'])
data2['Breast'] = encoder_breast
```

```python
data2 = createDummies(data2, "Breast_quad")
data2.head()
```

```python
data2.info()
```

```python
plt.hist(data2['Irradiat'])
```

```python
cols = ['Class', 'Age', 'Menopause', 'Tumor_size', 'Inv_nodes', 'Node_caps', 'Deg_malig', 'Breast', 'Breast_quad_central',
        'Breast_quad_left_low', 'Breast_quad_left_up', 'Breast_quad_right_low', 'Breast_quad_right_up']
```

```python
predictors = data2[cols]
target = data2['Irradiat']
```

```python
y_target = data2['Irradiat'].values
x_features_one = predictors.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_features_one, y_target, test_size = 0.20, random_state = 0)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
mlp.fit(X_train, y_train)
predictions = mlp.predict(X_test)
```
